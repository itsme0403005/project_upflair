```python
label_dict = {
    'rice':0,
    'maize':1,
    'chickpea':2,
    'kidneybeans':3,
    'pigeonpeas':4,
    'mothbeans':5,
    'mungbean':6,
    'blackgram':7,  
    'lentil':8,
    'pomegranate':9,
    'banana':10,
    'mango':11,
    'grapes':12,
    'watermelon':13,
    'muskmelon':14,
    'apple':15,
    'orange':16,
    'papaya':17,
    'coconut':18,
    'cotton':19,
    'jute':20,
    'coffee':21
}
df['label'] = df['label'].map(label_dict)
```

```python
df.isnull().sum()
```

```python
x=df.drop('label',axis=1)
y=df['label']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rndm_obj=RandomForestClassifier()
```

```python
rndm_obj.fit(x_train,y_train)
```

```python
rndm_obj.score(x_train,y_train)
```

```python
rndm_obj.score(x_test,y_test)
```
